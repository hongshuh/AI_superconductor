# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('./structure/data.csv')
data.columns=['material_id','composition','tc']
data
```

```python
train,test_val = train_test_split(data,train_size = 0.6)
val,test = train_test_split(test_val,test_size=0.5)
train.to_csv('./roost/data1/train.csv',index=None)
val.to_csv('./roost/data1/val.csv',index=None)
test.to_csv('./roost/data1/test.csv',index=None)
```
