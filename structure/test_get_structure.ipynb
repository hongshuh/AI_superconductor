```python
import pandas as pd
import json
from pymatgen.core import Structure
from pymatgen.core import composition
from pymatgen.io.cif import CifWriter
from sklearn.model_selection import train_test_split
```

```python
old_element = pd.read_csv('../uci_supercondutor_data/unique_m.csv')
old_element = old_element.iloc[:,-2:]
new_data = old_element.groupby('material').mean().reset_index()
new_data
```

```python
from pymatgen.ext.matproj import MPRester 
with MPRester("JJD0LmGd81VI8Xf03tliCN34Nm0RnQwl") as m:
    for x in range(3):
        test_structure = m.get_structures(new_data['material'][x])
        print(test_structure)
```

```python
from pymatgen.ext.matproj import MPRester 
with MPRester("JJD0LmGd81VI8Xf03tliCN34Nm0RnQwl") as m:
    i = 0
    materials = []
    critical_t = []
    for x in range(len(new_data)):
        test_structure = m.get_structures(new_data['material'][x])
        if len(test_structure) == 1:
            c = CifWriter(test_structure[0])
            c.write_file(f'{i}.cif')
            materials.append(new_data['material'][x])
            i+=1
            critical_t.append(new_data['critical_temp'][x])
```

```python
from pymatgen.ext.matproj import MPRester 
with MPRester("JJD0LmGd81VI8Xf03tliCN34Nm0RnQwl") as m:
    i = 300
    for x in range(5183,len(new_data)):
        test_structure = m.get_structures(new_data['material'][x])
        if len(test_structure) == 1:
            c = CifWriter(test_structure[0])
            c.write_file(f'{i}.cif')
            materials.append(new_data['material'][x])
            i+=1
            critical_t.append(new_data['critical_temp'][x])
```
